```python
import pandas as pd
filename = 'mh-cld-all_years.pkl'
df = pd.read_pickle(filename)
df

group_states = df['MH1'] 

# series of counts of each state type
ser = group_states.value_counts()
# make the series a df
df1 = ser.to_frame()
# replace the index to be numbers and state type is a column
df1['Disorder'] = df1.index.tolist()
numb_lst = list(range(0, len(df1.index.tolist())))
df1.set_index([pd.Index(numb_lst)], inplace = True)

df1
```

```python
df_disorders = pd.read_csv('Disorders_Key.csv')
df_disorders.head()
```

```python
disorder_dict = dict(zip(df_disorders.ID, df_disorders.DISORDER))
df1 = df1.replace({'Disorder': disorder_dict})
df1.head()
```

```python
df1.to_csv("./disorder_counts-all_years.csv")
```
