```python
import pandas as pd

filename1 = 'mhcld-puf-2013-csv.csv'
df2013 = pd.read_csv(filename1)
```

```python
df2013.head()
```

```python
filename2 = 'mhcld-puf-2014-csv.csv'
df2014 = pd.read_csv(filename2)

filename3 = 'mhcld-puf-2015-csv.csv'
df2015 = pd.read_csv(filename3)

filename4 = 'mhcld-puf-2016-csv.csv'
df2016 = pd.read_csv(filename4)

filename5 = 'mhcld-puf-2017-csv.csv'
df2017 = pd.read_csv(filename5)

filename6 = 'mhcld-puf-2018-csv.csv'
df2018 = pd.read_csv(filename6)

filename7 = 'mhcld-puf-2019-csv.csv'
df2019 = pd.read_csv(filename7)
```

```python
frames = [df2013, df2014, df2015, df2016, df2017, df2018, df2019]
dfAllYears = pd.concat(frames)
```

```python
dfAllYears
```

```python
dfAllYears = dfAllYears.dropna()
```

```python
#when downloading use a pickle (.pkl)
dfAllYears.to_pickle("./mh-cld-all_years.pkl")
```
